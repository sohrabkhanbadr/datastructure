## مرتب‌سازی درجی: نظم بخشیدن گام به گام

**مقدمه**

مرتب‌سازی درجی، الگوریتمی ساده و بصری برای مرتب‌سازی داده‌ها است. این الگوریتم از روشی شبیه به مرتب کردن کارت‌های بازی در دست استفاده می‌کند و به دلیل سادگی و کارایی، به یکی از الگوریتم‌های محبوب در علوم کامپیوتر تبدیل شده است.

**نحوه عملکرد:**

1. **بررسی گام به گام:** الگوریتم از دومین عنصر آرایه شروع می‌کند و آن را با عنصر قبلی خود مقایسه می‌کند.

2. **جابجایی:** اگر عنصر فعلی کوچکتر از عنصر قبلی باشد، جای آنها عوض می‌شود.

3. **تکرار:** این فرآیند برای عناصر بعدی آرایه تکرار می‌شود تا زمانی که به عنصر آخر برسیم.

**مزایای مرتب‌سازی درجی:**

* **سادگی:** الگوریتم مرتب‌سازی درجی به سادگی قابل فهم و پیاده‌سازی است.
* **کارایی:** در آرایه‌های کوچک و تقریباً مرتب‌شده، مرتب‌سازی درجی از الگوریتم‌های
    دیگر مانند مرتب‌سازی حبابی یا انتخابی سریع‌تر عمل می‌کند.
* **پایداری:** ترتیب عناصر با مقادیر برابر در آرایه ورودی پس از مرتب‌سازی حفظ می‌شود.
* **مصرف حافظه کم:** این الگوریتم به حافظه اضافی کمی نیاز دارد.

**معایب مرتب‌سازی درجی:**

* **عملکرد ضعیف در آرایه‌های بزرگ:** در آرایه‌های بزرگ، مرتب‌سازی درجی از
    الگوریتم‌های
    دیگر مانند مرتب‌سازی ادغام یا سریع کندتر عمل می‌کند.
* **جابجایی‌های زیاد:** در آرایه‌های نامرتب، مرتب‌سازی درجی نیاز به جابجایی‌های
    زیادی دارد که
    می‌تواند
    عملکرد آن را کندتر کند.

**نمونه کد:**

```
def insertion_sort(array):
    for i in range(1, len(array)):
        current_element = array[i]
        j = i - 1
        while j >= 0 and array[j] > current_element:
            array[j + 1] = array[j]
            j -= 1
        array[j + 1] = current_element
    return array
```

```
insertion_sort([7,3,6,1])
```

**پیچیدگی زمانی:**

پیچیدگی زمانی مرتب‌سازی درجی در بدترین حالت O(n^2) است.
